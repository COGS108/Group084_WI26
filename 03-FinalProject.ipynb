## Abstract
This project examines whether UC San Diego (UCSD) transfer majors with higher alumni earnings also exhibit greater admission selectivity. We wanted to explore whether stronger post-graduation salary outcomes were associated with higher admitted GPA midpoints and lower admit rates across majors, since students often choose majors with future career opportunities in mind.

To study this, we use major‑by‑year data from 2018, 2020, 2022, 2024, and combine UCSD's transfer admission indicators, admitted GPA midpoints and ranges, admit rates, applicant counts, and yield rates, with corresponding UCSD alumni earnings and its' percentiles to evaluate how academic competitiveness aligns with labor market outcomes.
We then analyze the relationship between earnings and selectivity through descriptive comparisons, trend visualizations, correlation analysis, and OLS regression models with controls for year, major fixed differences, and applicant volume. 

We found that, across all four years, engineering majors, especially for mechanical engineering and computer engineering, consistently emerge as the most selective, showing the highest GPA midpoints, the lowest admit rates, and pronounced upward trends in academic competitiveness. Arts and humanities majors, in contrast, maintain lower GPA thresholds and higher admit rates, reflecting more open admission structures.  These hierarchical patterns closely mirror the alumni earnings distribution, where engineering and technical fields rank at the top, mid‑STEM disciplines occupy the middle, and humanities and arts fields exhibit lower but more variable earnings profiles. Correlation and regression results show a positive association between median earnings and GPA midpoint and a significant negative association between earnings and admit rate, with the income–admit‑rate relationship strengthening in later years. This suggests that high‑earning majors are becoming increasingly selective primarily by reducing admission rates rather than raising GPA thresholds, which remain compressed near the upper bound of the 4.0 scale. 

Our findings also refine existing literature by showing that some “pre‑professional” majors, such as human biology, maintain high GPA thresholds despite moderate early‑career earnings, indicating that admission difficulty is shaped not only by immediate salary outcomes but also by long‑term academic and professional pathways. While earnings predict tightening admit rates, they do not strongly predict changes in GPA midpoints once persistent major‑level differences are accounted for. The results highlight a meaningful but nuanced link between labor‑market outcomes and transfer‑admission selectivity at UC San Diego. Limitations include aggregated data, GPA ceiling effects, small sample sizes for some majors, and reliance on early‑career earnings, suggesting opportunities for future work using longitudinal and applicant‑level datasets.

## Background and Prior Work


Choosing a university major is a crucial decision that impacts a student's academic trajectory and future career prospects. In recent years, there has been increasing attention to the "return on investment" of higher education, with prospective students often weighing the academic difficulty of a major against its potential graduate earnings. For transfer students—who typically have more defined career goals upon entering university—the competitiveness of admission (measured by the GPA distribution of admitted students (25th, 50th, and 75th percentiles)) is a primary evaluation criterion. This study uses the University of California, San Diego (UCSD) as a case study, a leading public research university where transfer admission selectivity and GPA thresholds vary significantly across different disciplines.

Previous research has confirmed a significant but nuanced link between university admission difficulty and post-graduation earnings. Data from the University of California Information Center suggests this relationship is particularly pronounced at highly competitive institutions like UCSD, especially in fields such as mechanical engineering and computer science, where UC alumni consistently rank among the highest earners five years after graduation. On the other hand, some "pre-professional" majors, such as human biology, maintain high GPA admission thresholds despite potentially lower undergraduate earnings compared to technical fields. This is often because these students go on to pursue advanced medical degrees, indicating that "admission difficulty" is not solely determined by current average salaries but also by long-term academic value.

Labor market conditions and alumni salary signals have also been shown to influence student demand, and consequently, the academic standards required for admission. It is widely believed that students tend to gravitate towards majors perceived to have higher economic returns, such as STEM fields and applied social sciences like business psychology. This shift in demand leads to increased admission difficulty and higher GPA admission thresholds. For example, UCSD's computer science program consistently maintains higher GPA requirements than humanities disciplines like history, directly reflecting its leading position in alumni earnings data.

Despite these observations, there is a lack of localized research specifically examining the year-to-year changes in UCSD's admission GPA distribution and its correlation with real-time average salary trends for specific majors. While general research suggests that a 10% increase in expected salary increases the likelihood of students choosing a particular major by nearly 14% to 18%, it remains unclear whether these enrollment preferences immediately translate into quantifiable changes in the 25th, and 75th percentiles of matriculation GPA. This project aims to bridge the gap between educational enrollment data and labor market wage signals by analyzing fifteen different majors at UCSD, spanning fields from history to mechanical engineering.

```python
# Import necessary libraries for data handling, visualization, and statistics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import re
import scipy.stats as stats
import os

# Load dataset if not already in memory
# If processed file exists, read it; otherwise build it using helper function
if "df_major" not in globals():
    proc_path = "data/02-processed/major_processed.csv"
    if os.path.exists(proc_path):
        df_major = pd.read_csv(proc_path, encoding="utf-8-sig")
    else:
        df_major = build_df_major(save=True)

# Create a copy to avoid modifying original dataset
df_major_flag = df_major.copy()

# List to store validation rules
rules = []

# Rule 1: Check if year values are valid
rules.append(("bad_year", ~df_major_flag["year"].isin([2018, 2020, 2022, 2024])))

# Rule 2: Admission rate must be between 0 and 1
if "admit_rate" in df_major_flag.columns:
    rules.append(("bad_admit_rate", (df_major_flag["admit_rate"] < 0) | (df_major_flag["admit_rate"] > 1)))

# Rule 3: Yield rate must be between 0 and 1
if "yield_rate" in df_major_flag.columns:
    rules.append(("bad_yield_rate", (df_major_flag["yield_rate"] < 0) | (df_major_flag["yield_rate"] > 1)))

# Rule 4: Counts should not be negative
for c in ["applicants", "admits", "enrollees"]:
    if c in df_major_flag.columns:
        rules.append((f"neg_{c}", df_major_flag[c] < 0))

# Rule 5: Logical consistency checks
if set(["applicants","admits"]).issubset(df_major_flag.columns):
    rules.append(("admits_gt_applicants", df_major_flag["admits"] > df_major_flag["applicants"]))

if set(["enrollees","admits"]).issubset(df_major_flag.columns):
    rules.append(("enrollees_gt_admits", df_major_flag["enrollees"] > df_major_flag["admits"]))

# Rule 6: GPA should be within reasonable bounds
if "gpa_mid" in df_major_flag.columns:
    rules.append(("bad_gpa_mid", (df_major_flag["gpa_mid"] < 0) | (df_major_flag["gpa_mid"] > 4.5)))

# Function to detect outliers using IQR method
def flag_iqr_outliers(s, k=1.5):
    s = s.dropna()
    if len(s) < 6:
        return pd.Series(False, index=s.index)  # Not enough data to compute outliers
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    lo, hi = q1 - k * iqr, q3 + k * iqr
    return (s < lo) | (s > hi)

# Columns to check for statistical outliers
iqr_cols = [c for c in ["applicants","admits","enrollees","gpa_mid","admit_rate","yield_rate"] if c in df_major_flag.columns]

# Apply IQR outlier detection within each year group
for col in iqr_cols:
    mask = (
        df_major_flag.groupby("year", group_keys=False)[col]
        .apply(lambda s: flag_iqr_outliers(s, k=1.5))
    )
    df_major_flag[f"out_{col}_iqr"] = False
    df_major_flag.loc[mask.index, f"out_{col}_iqr"] = mask

# Initialize column to store reasons for suspicious rows
df_major_flag["suspicious_reasons"] = ""

# Apply rule-based flags
for name, cond in rules:
    df_major_flag.loc[cond.fillna(False), "suspicious_reasons"] += f"{name};"

# Add outlier-based flags
for col in iqr_cols:
    flag_col = f"out_{col}_iqr"
    df_major_flag.loc[df_major_flag[flag_col] == True, "suspicious_reasons"] += f"{flag_col};"

# Final boolean flag indicating whether row is suspicious
df_major_flag["is_suspicious"] = df_major_flag["suspicious_reasons"].str.len() > 0

# Output summary and preview suspicious rows
print("Suspicious rows (major):", df_major_flag["is_suspicious"].sum())
display(df_major_flag[df_major_flag["is_suspicious"]].head(30))
```

#### Tidy and Clean Data

```python
# basic libraries for handling data
import os, re
import numpy as np
import pandas as pd

# set up file paths
RAW_DIR  = "data/00-raw"
PROC_DIR = "data/02-processed"
os.makedirs(PROC_DIR, exist_ok=True)

major_path = os.path.join(RAW_DIR, "2018-2024 Major Table.csv")

# load the raw csv file
df_major_raw = pd.read_csv(
    major_path,
    encoding="utf-8-sig",
    sep=",",
    engine="python"
)

# quick look at raw data
print("Major raw shape:", df_major_raw.shape)
display(df_major_raw.head())


# -------- helper functions --------

def to_numeric_percent(x):
    # turn things like '75%' into 0.75
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    if s == "":
        return np.nan
    s = s.replace("%", "").strip()
    try:
        v = float(s)
    except:
        return np.nan
    
    # sometimes values are like 75 instead of 0.75
    if v > 1:
        v = v / 100
    return v


def parse_gpa_range(x):
    # split GPA range like "3.4 - 3.9" into two numbers
    if pd.isna(x):
        return (np.nan, np.nan)
    s = str(x).strip()
    if s == "":
        return (np.nan, np.nan)

    # handle different dash types
    s = s.replace("–", "-").replace("—", "-")

    nums = re.findall(r"\d+(?:\.\d+)?", s)
    if len(nums) >= 2:
        lo = float(nums[0])
        hi = float(nums[1])
        if lo > hi:
            lo, hi = hi, lo
        return (lo, hi)
    
    return (np.nan, np.nan)


def standardize_major_name(x):
    # clean up spacing in major names
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    s = re.sub(r"\s+", " ", s)
    return s


# -------- rename columns --------

# make column names consistent and easier to use
rename_map = {
    "Year": "year",
    "Major": "major",
    "Applicants": "applicants",
    "Admits": "admits",
    "Enrollees": "enrollees",
    "Admit GPA range": "admit_gpa_range",
    "Enrollee GPA range": "enrollee_gpa_range",
    "Admit rate": "admit_rate",
    "Yield rate": "yield_rate",
}

df_major = df_major_raw.rename(columns=rename_map).copy()

# remove weird spaces if any
df_major.columns = df_major.columns.str.strip()

# only keep columns we care about
base_cols = [c for c in rename_map.values() if c in df_major.columns]
df_major = df_major[base_cols].copy()


# -------- cleaning --------

# convert year to numeric
if "year" in df_major.columns:
    df_major["year"] = pd.to_numeric(df_major["year"], errors="coerce").astype("Int64")

# clean major names
if "major" in df_major.columns:
    df_major["major"] = df_major["major"].apply(standardize_major_name)

# convert counts to numbers
for c in ["applicants", "admits", "enrollees"]:
    if c in df_major.columns:
        df_major[c] = pd.to_numeric(df_major[c], errors="coerce")

# convert percentages
for c in ["admit_rate", "yield_rate"]:
    if c in df_major.columns:
        df_major[c] = df_major[c].apply(to_numeric_percent)


# -------- tidy step (important) --------

# split GPA range into usable numeric columns
if "admit_gpa_range" in df_major.columns:
    lo_hi = df_major["admit_gpa_range"].apply(parse_gpa_range)

    df_major["gpa_low"] = lo_hi.apply(lambda t: t[0])
    df_major["gpa_high"] = lo_hi.apply(lambda t: t[1])

    # average GPA (easier for analysis later)
    df_major["gpa_mid"] = (df_major["gpa_low"] + df_major["gpa_high"]) / 2

    # spread of GPA range
    df_major["gpa_width"] = df_major["gpa_high"] - df_major["gpa_low"]


# -------- filter + final shape --------

# only keep years we want
YEARS = [2018, 2020, 2022, 2024]
if "year" in df_major.columns:
    df_major = df_major[df_major["year"].isin(YEARS)].copy()

# drop rows missing key info
df_major = df_major.dropna(subset=["year", "major"]).copy()

# reorder columns to make it cleaner
final_cols = [
    "year", "major",
    "applicants", "admits", "enrollees",
    "admit_rate", "yield_rate",
    "admit_gpa_range", "enrollee_gpa_range",
    "gpa_low", "gpa_high", "gpa_mid", "gpa_width",
]

final_cols = [c for c in final_cols if c in df_major.columns]
df_major = df_major[final_cols].copy()


# -------- quick checks --------

print("Major cleaned shape:", df_major.shape)
display(df_major.head(10))

# check missing values
miss = df_major.isna().mean().sort_values(ascending=False)
print("Major missingness (fraction):")
display(miss)

print("Has gpa_mid?", "gpa_mid" in df_major.columns)
if "gpa_mid" in df_major.columns:
    print("gpa_mid missing:", df_major["gpa_mid"].isna().mean())


# -------- save result --------

major_out = os.path.join(PROC_DIR, "major_processed.csv")
df_major.to_csv(major_out, index=False, encoding="utf-8-sig")

print("Saved:", major_out)
```

```python
if "df_earn" not in globals():

    # use processed file first if it already exists
    proc_path = "data/02-processed/earnings_processed.csv"
    raw_path  = "data/00-raw/UC alumni at work.csv"

    if os.path.exists(proc_path):
        df_earn = pd.read_csv(proc_path, encoding="utf-8-sig")
        print("Loaded df_earn from processed:", proc_path)

    elif os.path.exists(raw_path):
        df_earn_raw = pd.read_csv(raw_path, encoding="utf-8-sig", sep=",", engine="python")
        print("Loaded df_earn from raw:", raw_path)

        # clean column names first
        df_earn_raw.columns = df_earn_raw.columns.str.strip()

        # rename columns to shorter names
        rename_map = {
            "Year": "year",
            "Major": "major",
            "Median Annual Earnings": "earn_med",
            "Percentile (25) of Annual Earnings": "earn_p25",
            "Percentile (75) of Annual Earnings": "earn_p75",
            "Earners Count": "earners_count",
            "Count of Number of Records": "records_count",
        }
        df_earn = df_earn_raw.rename(
            columns={k: v for k, v in rename_map.items() if k in df_earn_raw.columns}
        ).copy()

        def normalize_year(x):
            # some year values show up as ranges like 2022-2023
            # in that case we keep the second year
            if pd.isna(x):
                return np.nan
            s = str(x).strip()
            m = re.match(r"^\s*(\d{4})\s*-\s*(\d{4})\s*$", s)
            if m:
                return int(m.group(2))
            try:
                return int(float(s))
            except:
                return np.nan

        if "year" in df_earn.columns:
            df_earn["year"] = df_earn["year"].apply(normalize_year).astype("Int64")

    else:
        raise FileNotFoundError(
            "Cannot find processed or raw earnings file. Run get_data.get_raw(...) first."
        )

# work on a copy so the original table stays unchanged
df_earn_flag = df_earn.copy()
df_earn_flag.columns = df_earn_flag.columns.str.strip()

# just in case the year column was not renamed earlier
if "year" not in df_earn_flag.columns and "Year" in df_earn_flag.columns:
    df_earn_flag = df_earn_flag.rename(columns={"Year": "year"})

rules = []

# year should only be one of the project years
if "year" in df_earn_flag.columns:
    rules.append(("bad_year", ~df_earn_flag["year"].isin([2018, 2020, 2022, 2024])))
else:
    print("Warning: no 'year' column found; skipping bad_year rule.")

# earnings should not be negative or unrealistically large
for c in ["earn_p25", "earn_med", "earn_p75"]:
    if c in df_earn_flag.columns:
        rules.append((f"neg_{c}", df_earn_flag[c] < 0))
        rules.append((f"too_big_{c}", df_earn_flag[c] > 1_000_000))

# basic ordering check: p25 <= median <= p75
if set(["earn_p25", "earn_med"]).issubset(df_earn_flag.columns):
    rules.append(("p25_gt_med", df_earn_flag["earn_p25"] > df_earn_flag["earn_med"]))
if set(["earn_med", "earn_p75"]).issubset(df_earn_flag.columns):
    rules.append(("med_gt_p75", df_earn_flag["earn_med"] > df_earn_flag["earn_p75"]))

# counts should not be negative
for c in ["earners_count", "records_count"]:
    if c in df_earn_flag.columns:
        rules.append((f"neg_{c}", df_earn_flag[c] < 0))

def flag_iqr_outliers(s, k=1.5):
    # standard IQR rule for possible outliers
    s = s.dropna()
    if len(s) < 6:
        return pd.Series(False, index=s.index)
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    lo, hi = q1 - k * iqr, q3 + k * iqr
    return (s < lo) | (s > hi)

# columns used for outlier checks
iqr_cols = [c for c in ["earn_p25", "earn_med", "earn_p75", "earners_count"] if c in df_earn_flag.columns]

# check outliers within each year so different years are compared fairly
if "year" in df_earn_flag.columns:
    for col in iqr_cols:
        mask = (
            df_earn_flag.groupby("year", group_keys=False)[col]
            .apply(lambda s: flag_iqr_outliers(s, k=1.5))
        )
        df_earn_flag[f"out_{col}_iqr"] = False
        df_earn_flag.loc[mask.index, f"out_{col}_iqr"] = mask
else:
    print("Warning: no 'year' column found; skipping IQR-by-year outliers.")

# collect all reasons in one text column
df_earn_flag["suspicious_reasons"] = ""
for name, cond in rules:
    df_earn_flag.loc[cond.fillna(False), "suspicious_reasons"] += f"{name};"

# add IQR outlier flags to the same reason column
for col in iqr_cols:
    flag_col = f"out_{col}_iqr"
    if flag_col in df_earn_flag.columns:
        df_earn_flag.loc[df_earn_flag[flag_col] == True, "suspicious_reasons"] += f"{flag_col};"

# final flag for rows that triggered at least one rule
df_earn_flag["is_suspicious"] = df_earn_flag["suspicious_reasons"].str.len() > 0

print("Suspicious rows (earnings):", df_earn_flag["is_suspicious"].sum())
display(df_earn_flag[df_earn_flag["is_suspicious"]].head(30))
```

```python
# import basic libraries
import os, re
import numpy as np
import pandas as pd

# set up paths
RAW_DIR = "data/00-raw"
PROC_DIR = "data/02-processed"
os.makedirs(PROC_DIR, exist_ok=True)

earn_path = os.path.join(RAW_DIR, "UC alumni at work.csv")

# load raw earnings data
df_earn_raw = pd.read_csv(
    earn_path,
    encoding="utf-8-sig",
    sep=",",
    engine="python",
)

print("Earnings raw shape:", df_earn_raw.shape)
display(df_earn_raw.head())


# -------- helper functions --------

def standardize_major_name(x):
    # clean up spacing in major names
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    s = re.sub(r"\s+", " ", s)
    return s


def to_numeric_money(x):
    # convert money strings like "$120,548" into numbers
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    if s == "":
        return np.nan
    s = s.replace("$", "").replace(",", "")
    try:
        return float(s)
    except:
        return np.nan


# -------- rename columns --------

# make column names shorter and easier to work with
rename_map = {
    "Year": "year",
    "Major": "major",
    "Median Annual Earnings": "earn_med",
    "Percentile (25) of Annual Earnings": "earn_p25",
    "Percentile (75) of Annual Earnings": "earn_p75",
    "Earners Count": "earners_count",
    "Count of Number of Records": "records_count",
}

df_earn = df_earn_raw.rename(columns=rename_map).copy()

# keep only columns we actually need
keep_cols = [c for c in rename_map.values() if c in df_earn.columns]
df_earn = df_earn[keep_cols].copy()


# -------- cleaning --------

def normalize_year(x):
    # handle cases like "2022-2023" by taking the second year
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    m = re.match(r"^\s*(\d{4})\s*-\s*(\d{4})\s*$", s)
    if m:
        return int(m.group(2))
    try:
        return int(float(s))
    except:
        return np.nan

# convert year and major
df_earn["year"] = df_earn["year"].apply(normalize_year).astype("Int64")
df_earn["major"] = df_earn["major"].apply(standardize_major_name)

# convert earnings columns to numeric
for c in ["earn_med", "earn_p25", "earn_p75"]:
    if c in df_earn.columns:
        df_earn[c] = df_earn[c].apply(to_numeric_money)

# convert count columns
for c in ["earners_count", "records_count"]:
    if c in df_earn.columns:
        df_earn[c] = pd.to_numeric(df_earn[c], errors="coerce")


# -------- filtering --------

# keep only the years used in our project
YEARS = [2018, 2020, 2022, 2024]
df_earn = df_earn[df_earn["year"].isin(YEARS)].copy()

# remove rows missing key fields
df_earn = df_earn.dropna(subset=["year", "major"]).copy()


# -------- quick checks --------

print("Earnings cleaned shape:", df_earn.shape)
display(df_earn.head(10))

# check missing values
miss = df_earn.isna().mean().sort_values(ascending=False)
print("Earnings missingness (fraction):")
display(miss)


# -------- save cleaned file --------

earn_out = os.path.join(PROC_DIR, "earnings_processed.csv")
df_earn.to_csv(earn_out, index=False, encoding="utf-8-sig")

print("Saved:", earn_out)
```

#### Relationship Between Median Earnings, GPA Midpoint, and Admit Rate

```python
# calculate correlation results between two variables
# we report both Pearson and Spearman to compare linear and rank-based patterns
def corr_report(df, x, y):
    sub = df.dropna(subset=[x, y]).copy()
    n = len(sub)

    # not enough data to compute correlation
    if n < 3:
        return {
            "n": n,
            "pearson_r": np.nan,
            "pearson_p": np.nan,
            "spearman_rho": np.nan,
            "spearman_p": np.nan
        }

    pear_r, pear_p = stats.pearsonr(sub[x], sub[y])
    spr_r, spr_p = stats.spearmanr(sub[x], sub[y])

    return {
        "n": n,
        "pearson_r": pear_r,
        "pearson_p": pear_p,
        "spearman_rho": spr_r,
        "spearman_p": spr_p
    }


# variable pairs we want to compare with median earnings
pairs = [
    ("gpa_mid", "earn_med"),
    ("admit_rate", "earn_med"),
    ("gpa_width", "earn_med"),
    ("yield_rate", "earn_med"),
]

# print correlation results for each pair
for x, y in pairs:
    print(f"{x} vs {y}:", corr_report(df, x, y))
```

Here we calculate correlation statistics to measure the relationship between alumni earnings and several admissions-related variables, including GPA midpoint, admit rate, GPA width, and yield rate. Correlation analysis is useful here because it provides a quick glance at whether majors with stronger labor-market outcomes also tend to be more selective in transfer admissions. We report both Pearson and Spearman correlations so that we can compare linear relationships as well as broader rank-based patterns. This step is important for our project because it helps identify which admissions measures are most strongly associated with earnings before moving on to more detailed regression analysis.

#### Constructing the Δ (Change) Dataset for Trend Analysis (2018, 2020, 2022, 2024)

```python
# keep only the years we want for the change analysis
years_keep = [2018, 2020, 2022, 2024]

df4 = df.copy()
df4["year"] = pd.to_numeric(df4["year"], errors="coerce").astype("Int64")

# keep rows with the key variables we need
df4 = df4[
    df4["year"].isin(years_keep)
].dropna(subset=["major", "gpa_mid", "admit_rate", "earn_med"]).copy()

df4["year"] = df4["year"].astype(int)

print("Years in df4:", sorted(df4["year"].unique().tolist()))

# reshape to wide format so each major has one row
w_gpa = df4.pivot_table(index="major", columns="year", values="gpa_mid", aggfunc="mean")
w_adm = df4.pivot_table(index="major", columns="year", values="admit_rate", aggfunc="mean")

print("Pivot GPA columns:", w_gpa.columns.tolist())
print("Pivot Admit columns:", w_adm.columns.tolist())

# start a new table with one row per major
delta = pd.DataFrame({"major": w_gpa.index})

# add GPA and admit rate values for each year
for y in years_keep:
    delta[f"gpa_mid_{y}"] = w_gpa[y].values
    delta[f"admit_rate_{y}"] = w_adm[y].values

# compute step-by-step GPA changes
delta["d_gpa_20_18"] = delta["gpa_mid_2020"] - delta["gpa_mid_2018"]
delta["d_gpa_22_20"] = delta["gpa_mid_2022"] - delta["gpa_mid_2020"]
delta["d_gpa_24_22"] = delta["gpa_mid_2024"] - delta["gpa_mid_2022"]

# compute step-by-step admit rate changes
delta["d_adm_20_18"] = delta["admit_rate_2020"] - delta["admit_rate_2018"]
delta["d_adm_22_20"] = delta["admit_rate_2022"] - delta["admit_rate_2020"]
delta["d_adm_24_22"] = delta["admit_rate_2024"] - delta["admit_rate_2022"]

# also compute overall change from 2018 to 2024
delta["d_gpa_24_18"] = delta["gpa_mid_2024"] - delta["gpa_mid_2018"]
delta["d_adm_24_18"] = delta["admit_rate_2024"] - delta["admit_rate_2018"]

# add earnings measures for later comparison
earn_level_avg = (
    df4.groupby("major", as_index=False)["earn_med"]
    .mean()
    .rename(columns={"earn_med": "earn_med_avg_18_24"})
)

earn_2024 = (
    df4[df4["year"] == 2024][["major", "earn_med"]]
    .rename(columns={"earn_med": "earn_med_2024"})
)

delta = delta.merge(earn_level_avg, on="major", how="left").merge(earn_2024, on="major", how="left")

display(delta)
```

Here we construct a change-based dataset to focus on how majors shift over time rather than only comparing their levels in a single year. This step is useful because our earlier analyses mainly examined static relationships between earnings and selectivity, while the delta dataset allows us to measure whether increases or decreases in GPA midpoint and admit rate occur alongside changes in major-level earnings across the 2018–2024 period. By organizing the data into year-to-year and long-run differences, we can better study whether majors that become more selective also tend to show stronger earnings outcomes. #### Model Summary Table: Comparing the Effect of Median Earnings Across Regression Specifications

```python
# Extract key statistics from a fitted regression model
def summarize_model(model, term="earn_med"):
    # Coefficient (beta) for the variable of interest
    b = model.params.get(term, np.nan)
    
    # p-value for statistical significance
    p = model.pvalues.get(term, np.nan)
    
    # 95% confidence interval for the coefficient
    ci = model.conf_int().loc[term].tolist() if term in model.params.index else [np.nan, np.nan]
    
    # Model fit and sample size
    r2 = model.rsquared
    n = int(model.nobs)
    
    # Return results in a clean dictionary format
    return {
        "beta": b,
        "p": p,
        "ci_low": ci[0],
        "ci_high": ci[1],
        "R2": r2,
        "N": n
    }


# Compare effect of earnings across different model specifications
print("m3 gpa_mid ~ earn_med + year:", summarize_model(m3, "earn_med"))
print("m4 admit_rate ~ earn_med + year:", summarize_model(m4, "earn_med"))

print("m5 gpa_mid ~ earn_med + year + C(major):", summarize_model(m5, "earn_med"))
print("m6 admit_rate ~ earn_med + year + C(major):", summarize_model(m6, "earn_med"))

# Optional model with applicants control (if available)
if "m7" in globals():
    print("m7 admit_rate ~ earn_med + year + applicants + C(major):",
          summarize_model(m7, "earn_med"))
```

Here we create a model summary table to compare how the estimated effect of median earnings changes across different regression specifications. This is useful because the coefficient on earnings may look meaningful in a simpler model but weaken after adding controls such as year, major, and applicant volume. By placing the key results side by side, we can more easily evaluate the stability, significance, and confidence intervals of the earnings effect across models. This summary is important for our project because it helps show whether the relationship between earnings and transfer selectivity is robust or whether it depends heavily on the specification being used.

#### Building a Composite Selectivity Score (2018, 2020, 2022, 2024)

```python
# Years included in the composite score
YEARS = [2018, 2020, 2022, 2024]

# Standardize a series so values are comparable
def zscore(s: pd.Series) -> pd.Series:
    s = pd.to_numeric(s, errors="coerce")
    mu = s.mean()
    sd = s.std(ddof=0)

    # If all values are the same, return values centered at 0
    if sd == 0 or np.isnan(sd):
        return (s - mu)

    return (s - mu) / sd


# Compute a yearly selectivity score for each major
def compute_yearly_scores(df: pd.DataFrame) -> pd.DataFrame:
    d = df[df["year"].isin(YEARS)].copy()

    # Start with GPA midpoint and admit rate
    cols = ["gpa_mid", "admit_rate"]

    # Include earnings if available
    if "earn_med" in d.columns:
        cols.append("earn_med")

    # Convert selected columns to numeric and keep rows with the key variables
    d[cols] = d[cols].apply(pd.to_numeric, errors="coerce")
    d = d.dropna(subset=["year", "major", "gpa_mid", "admit_rate"]).copy()

    # Standardize values within each year
    d["z_gpa"] = d.groupby("year")["gpa_mid"].transform(zscore)

    # Lower admit rate means higher selectivity, so we reverse the sign
    d["z_admit_inv"] = -d.groupby("year")["admit_rate"].transform(zscore)

    # If earnings are available, include them in the composite score
    if "earn_med" in d.columns:
        d["z_earn"] = d.groupby("year")["earn_med"].transform(zscore)
        d["selectivity_score_year"] = d["z_gpa"] + d["z_admit_inv"] + d["z_earn"]
    else:
        d["selectivity_score_year"] = d["z_gpa"] + d["z_admit_inv"]

    # Return the main columns we want to keep
    return d[
        ["year", "major", "gpa_mid", "admit_rate"]
        + (["earn_med"] if "earn_med" in d.columns else [])
        + ["selectivity_score_year"]
    ]


# Compute yearly scores
yearly = compute_yearly_scores(df)

# Aggregate the score across all years for each major
agg = (
    yearly
    .groupby("major")
    .agg(
        years=("year", lambda x: sorted(x.unique().tolist())),
        n_years=("year", "nunique"),
        score_mean=("selectivity_score_year", "mean"),
        score_median=("selectivity_score_year", "median"),
    )
    .reset_index()
)

# Keep only majors that appear in all four years
agg_all4 = agg[agg["n_years"] == 4].sort_values("score_mean", ascending=False)

display(agg_all4.head(20))
```

Here we build a composite selectivity score to combine multiple indicators of competitiveness into a single summary measure. This is useful because GPA midpoint, admit rate, and median earnings each capture a different part of how selective or high-return a major may be, but looking at them separately can make the overall pattern harder to compare across majors. By standardizing each variable within year and combining them into one score, we can rank majors more consistently across the 2018–2024 period. This step is important for our project because it provides a broader way to evaluate which majors are consistently the most selective and economically advantaged when multiple indicators are considered together.

## Discusison and Conclusion

Our investigation into the relationship between the selectivity of transfer admissions at the University of California, San Diego (UCSD) in 2018, 2020, 2022, and 2024 and alumni income shows several important patterns. Engineering disciplines, particularly mechanical engineering, computer engineering, and bioengineering, consistently ranked highest in terms of GPA requirements and selectivity, forming a clear hierarchical structure that mirrors the income hierarchy observed in alumni salary data. Our regression analysis confirmed a statistically significant positive correlation between the median income and the GPA midpoint, indicating that majors leading to high-paying careers tend to admit transfer students with stronger academic backgrounds. We found a significant negative correlation between income and admission rates, with this relationship strengthening over time and becoming particularly evident in recent years. This suggests that the selectivity of high-paying majors is becoming increasingly stringent. Our analysis reveals that income does predict a decrease in admission rates, meaning that the proportion of applicants admitted to high-paying majors will decline, even if their GPA thresholds remain relatively stable.

Our research results have extended the existing literature in several meaningful ways. The previous research conducted by the Information Center of the University of California demonstrated that alumni from the university's mechanical engineering and computer science departments, among others, rose to the top income ranks five years after graduation. Our results confirm that these disciplines maintain the most competitive admission standards for transfer students. Our results also complicate the narrative found in sources that suggests a direct relationship between selectivity and income. In fact, some majors, such as human biology, characterized as “pre professional” majors, despite having moderate early career incomes, still maintain high GPA thresholds. This fact supports our observations in the background, that students may weigh long-term values (such as medical school admissions) rather than just the immediate salary prospects. Thus that "admission difficulty" is not dependent only on current labor market signals, but also on the perceived pathways to advanced degrees and professional schools.

We have come to understand that the relationship between professional selectivity and income is both real and subtle. The positive correlation between the GPA midpoint and income confirms that competitive academic programs often lead to high-paying careers, but the insignificant relationship between income and GPA, which varies over time, indicates that this hierarchical structure is relatively stable rather than responding dynamically to short-term salary fluctuations. The most striking finding is the increasingly significant negative correlation between income and admission rate, and this negative correlation. The pattern suggests that institutions are responding to the growing demand for high-income majors by admitting a smaller proportion of applicants. The differences in selectivity between different dimensions, GPA requirements and admission rates, are a key contribution of our work, revealing that the competition for high-income majors is exacerbated mainly by lowering the admission probability, not by raising GPA threshold.

Several important limitations here with our analysis. Our data was aggregated at the professional grade level, thus unable to analyze individual-level factors such as personal backgrounds which might impact the relationship between selectivity and income. The 4.0 GPA upper limit creates a hard limit, which might mask the improvement in the quality of top applicants, especially in highly competitive majors like mechanical engineering, where 75% are already at 3.98-4.00. Our income data only reflects the first two years after graduation, which might not reflect the long-term career trajectory and that it relies on alumni who agree to share their information. To be say, the two-year structure of our data captures medium-term trends but cannot reveal the yearly fluctuations that might be more consistent with the timing of student application decisions.
