## Discusison and Conclusion

Our investigation into the relationship between the selectivity of transfer admissions at the University of California, San Diego in 2018, 2020, 2022, and 2024 and alumni income shows several important patterns. Engineering disciplines, particularly mechanical engineering, computer engineering, and bioengineering, consistently ranked highest in terms of GPA requirements and selectivity, forming a clear hierarchical structure that mirrors the income hierarchy observed in alumni salary data. Our regression analysis confirmed a statistically significant positive correlation between the median income and the GPA midpoint, indicating that majors leading to high-paying careers tend to admit transfer students with stronger academic backgrounds. We found a significant negative correlation between income and admission rates, with this relationship strengthening over time and becoming particularly evident in recent years. This suggests that the selectivity of high-paying majors is becoming increasingly stringent. Our analysis reveals that income does predict a decrease in admission rates, meaning that the proportion of applicants admitted to high-paying majors will decline, even if their GPA thresholds remain relatively stable.

Our research results have extended the existing literature in several meaningful ways. The previous research conducted by the Information Center of the University of California demonstrated that alumni from the university's mechanical engineering and computer science departments, among others, rose to the top income ranks five years after graduation. Our results confirm that these disciplines maintain the most competitive admission standards for transfer students. Our results also complicate the narrative found in sources that suggests a direct relationship between selectivity and income. In fact, some majors, such as human biology, characterized as “pre professional” majors, despite having moderate early career incomes, still maintain high GPA thresholds. This fact supports our observations in the background, that students may weigh long-term values (such as medical school admissions) rather than just the immediate salary prospects. Thus that "admission difficulty" is not dependent only on current labor market signals, but also on the perceived pathways to advanced degrees and professional schools.

We have come to understand that the relationship between professional selectivity and income is both real and subtle. The positive correlation between the GPA midpoint and income confirms that competitive academic programs often lead to high-paying careers, but the insignificant relationship between income and GPA, which varies over time, indicates that this hierarchical structure is relatively stable rather than responding dynamically to short-term salary fluctuations. The most striking finding is the increasingly significant negative correlation between income and admission rate, and this negative correlation. The pattern suggests that institutions are responding to the growing demand for high-income majors by admitting a smaller proportion of applicants. The differences in selectivity between different dimensions, GPA requirements and admission rates, are a key contribution of our work, revealing that the competition for high-income majors is exacerbated mainly by lowering the admission probability, not by raising GPA threshold.

Several important limitations here with our analysis. Our data was aggregated at the professional grade level, thus unable to analyze individual-level factors such as personal backgrounds which might impact the relationship between selectivity and income. The 4.0 GPA upper limit creates a hard limit, which might mask the improvement in the quality of top applicants, especially in highly competitive majors like mechanical engineering, where 75% are already at 3.98-4.00. The sample size of some majors is also very small making the average income vulnerable to fluctuations from a small number of individuals. Our income data only reflects the first two years after graduation, which might not reflect the long-term career trajectory and that it relies on alumni who agree to share their information. To be say, the two-year structure of our data captures medium-term trends but cannot reveal the yearly fluctuations that might be more consistent with the timing of student application decisions.

If this investigation is to continue beyond the current project, several expansions would be valuable. By combining longitudinal data that tracks students from their enrollment through their early career outcomes, we can directly test whether individual-level GPA can predict income within the profession. Collecting data on the number of applicants will also help determine whether the changes in selectivity are driven by an increase in demand or a decrease in supply. Extending the time frame to income 5 to 10 years after graduation will better capture the long-term returns of professions that require graduate education. Qualitative studies of transfer students can clarify whether students truly understand and respond to alumni income data when making major decisions. Finally, comparative analyses with other California university campuses will determine if our findings reflect system-wide patterns or dynamics that is only unique to UCSD.

Throughout the entire project, we have always maintained an awareness of the ethical dimension of the analysis. We used aggregated and de-identified data to minimize privacy issues to the greatest extent possible, but we recognized that income and enrollment statistics might reflect and could perpetuate structural inequality. The observed correlation between professional income and selectivity might be misinterpreted as a reason for prioritizing certain research fields. However, this relationship is the result of complex factors, including examples like labor market discrimination or the historically insufficient funding for specific disciplines. We caution against using our research results to rank professions by value or guiding students into specific fields without considering their various forms of success beyond salary. Future work should explicitly examine whether the pathways to high-income professions are fair for all backgrounds of students, and whether the selectivity we observed reflects elite sorting or the accumulation of advantages across educational stages.
